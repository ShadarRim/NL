```python
def split_text(text: str, max_chars: int = 50) -> list:
    """
    Разделяет текст по словам (разделитель - пробел)
    :param text: Текст
    :param max_chars: Максимально допустимая длина части текста
    """
    result = []
    text_words = text.rstrip('\n').replace(u'\xa0', u' ').split(' ')
    temp = "".join(filter(str.isalnum, text_words[0])).lower()

    for word in text_words[1:]:
        #print(word)
        word = "".join(filter(str.isalnum, word)).lower()
        if word == " ":
            continue
        if len(temp + word) > max_chars:
            while temp[0]==" ":
                temp = temp[1:] 
            result.append(temp)
            temp = word
        else:
            temp += f' {word}'
    
    while temp[0]==" ":
        temp = temp[1:]

       
            
    result.append(temp)

    
    return result
```

```python
def ceaser(text: str, shift: int) -> str:
    """
    Зашифруем Цезарем
    :text: Шифрумеый текст
    :shift: Сдвиг
    """ 

    result = ''

    shift = shift % 32
    for elem in text:
       
        if 1040 <= ord(elem) <= 1071:
            sh_el = ord(elem)+shift 
            if sh_el > 1071:
                sh_el = 1040 + sh_el - 1071
            elif sh_el < 1040:
                sh_el = 1071 + 1040 - sh_el
        elif 1072 <= ord(elem) <= 1103:
            sh_el = ord(elem)+shift
            if sh_el > 1103:
                sh_el = 1072 + sh_el - 1103
            elif sh_el < 1072:
                sh_el = 1103 + 1072 - sh_el
        else:
            sh_el = ord(elem)  

        result = result + chr(sh_el)

    
    return result[:-1]
```

```python
result = []

with open("text.txt") as file:
    for item in file:
        for elem in split_text(item):
            if len(elem) > 42: 
                while "  " in elem:
                    elem = elem.replace("  ", " ") 
              
                result.append(elem)
    
pprint(result)
```

```python
CHARS = set('абвгдежзийклмнопрстуфхцчшщъыьэюя ')
INDEX_TO_CHAR = ['none'] + [w for w in CHARS]
CHAR_TO_INDEX = {w: i for i, w in enumerate(INDEX_TO_CHAR)}
```

```python
len(INDEX_TO_CHAR)
```

```python
def texttocode(text):
    MAX_LEN = 50
    X = torch.zeros((len(text), MAX_LEN), dtype=int)
    for i in range(len(text)):
        for j, w in enumerate(text[i]):
            if j >= MAX_LEN:
                break
            X[i, j] = CHAR_TO_INDEX.get(w, CHAR_TO_INDEX['none'])
    
    return X
```

```python
X = texttocode(cr_res_sh)
```

```python
Y = texttocode(res_sh)
```

```python
X[0:10]
```

```python
class Network(torch.nn.Module):
    def __init__(self):
        super(Network, self).__init__()
        self.embed = torch.nn.Embedding(len(CHAR_TO_INDEX), 34)
        self.rnn = torch.nn.RNN(34, 128, batch_first = True)
        self.liner = torch.nn.Linear(128, len(INDEX_TO_CHAR))
        
    def forward(self, sentences, state=None):
        embed = self.embed(sentences)
        o, s = self.rnn(embed)
        out = self.liner(o)
        return out
```

```python
model = Network()
```

```python
criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=.05)
```

```python
for ep in range(100):
    start = time.time()
    train_loss = 0.
    train_passed = 0

    for i in range(int(len(X) / 100)):
        batch_X = X[i * 100:(i + 1) * 100]
        batch_Y = Y[i * 100:(i + 1) * 100]
        X_batch = batch_X[:, :]
        Y_batch = batch_Y[:, :].flatten()

        optimizer.zero_grad()
        answers = model.forward(X_batch)
        answers = answers.view(-1, len(INDEX_TO_CHAR))
        loss = criterion(answers, Y_batch)
        train_loss += loss.item()

        loss.backward()
        optimizer.step()
        train_passed += 1

    print("Epoch {}. Time: {:.3f}, Train loss: {:.3f}".format(ep, time.time() - start, train_loss / train_passed))
    #generate_sentence()
```
